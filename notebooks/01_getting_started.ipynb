# Getting Started with polymarket-pandas

Every endpoint returns a **pandas DataFrame**. No API keys needed for public endpoints.

This notebook walks through the basics: searching markets, reading orderbooks,
fetching price history, and browsing events.

```python
!pip install -q polymarket-pandas matplotlib
```

```python
from polymarket_pandas import PolymarketPandas

client = PolymarketPandas()
```

## Search Markets

Fetch the top 20 open markets sorted by 24-hour volume.

```python
markets = client.get_markets(
    closed=False,
    order=["volume24hr"],
    ascending=False,
    limit=20,
)
markets[["slug", "question", "volume24hr", "endDate"]].head(10)
```

## Orderbook

Pick the first token from the results and inspect the live L2 orderbook.

```python
token_id = markets["clobTokenIds"].iloc[0]
print(f"Token ID: {token_id}")

book = client.get_orderbook(token_id)
book.head(10)
```

## Price History

Fetch daily price history for the past week and plot it.

```python
import matplotlib.pyplot as plt

history = client.get_price_history(market=token_id, interval="1w", fidelity=60)

plt.figure(figsize=(10, 4))
plt.plot(history["t"], history["p"], marker=".", linewidth=1)
plt.title(f"Price History — {markets['question'].iloc[0][:60]}")
plt.xlabel("Time (UTC)")
plt.ylabel("Price")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Midpoint & Spread

Get the current midpoint price and bid-ask spread for the same token.

```python
mid = client.get_midpoint_price(token_id)
spread = client.get_spread(token_id)

print(f"Midpoint: {mid:.4f}")
print(f"Spread:   {spread:.4f}")
```

## Events

Events group related markets together. Expanding markets inlines each
market's columns with a `markets` prefix.

```python
events = client.get_events(
    closed=False,
    expand_markets=True,
    limit=10,
)
events[["title", "marketsQuestion", "marketsOutcomePrices", "endDate"]].head(10)
```
