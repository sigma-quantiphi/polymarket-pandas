# Portfolio Tracking — Positions, Trades & Leaderboard

The **Data API** endpoints for positions and trades are public -- they
require a wallet address but **no authentication**. The leaderboard is
fully public with no address needed.

This notebook covers:
- Trader and builder leaderboards
- Public trade history for any market
- Position queries (with a wallet address)
- Market positions and top holders

```python
from polymarket_pandas import PolymarketPandas

client = PolymarketPandas()
```

## Leaderboard

The trader leaderboard is fully public. Sort by PNL, volume, or other metrics
over different time windows.

```python
lb = client.get_leaderboard(timePeriod="WEEK", orderBy="PNL", limit=25)
lb.head(15)
```

## Builder Leaderboard

Builders (liquidity providers / market makers) have their own leaderboard.

```python
builder_lb = client.get_builder_leaderboard(timePeriod="WEEK", limit=10)
builder_lb.head(10)
```

## Public Trade History

Fetch recent trades for any market -- no auth needed. We first grab a
high-volume market, then pull its trades.

```python
# Get a high-volume market to query trades for
markets = client.get_markets(
    closed=False,
    order=["volume24hr"],
    ascending=False,
    limit=5,
)

token_id = markets["clobTokenIds"].iloc[0]
condition_id = markets["conditionId"].iloc[0]
print(f"Market: {markets['question'].iloc[0][:80]}")
print(f"Token ID: {token_id}")
print(f"Condition ID: {condition_id}")
```

```python
trades = client.get_trades(market=[token_id], limit=50)
trades.head(15)
```

## Market Positions

See all traders holding positions in a specific market. The `market`
parameter here is a **token ID** (not a condition ID).

```python
market_positions = client.get_market_positions(market=token_id, limit=20)
market_positions.head(10)
```

## Top Holders

Fetch the largest token holders for a market.

```python
holders = client.get_top_holders(market=[token_id], limit=20)
holders.head(10)
```
